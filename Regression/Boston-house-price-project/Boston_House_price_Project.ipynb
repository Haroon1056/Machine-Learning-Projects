```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Import the Boston House Price Dataset

```python
dataset = pd.read_csv("Boston-house-price-data.csv")
```

#### Check the Correlation

```python
dataset.corr()
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(dataset.corr(), annot=True)
plt.show()
```

#### Splitting Feature and Target

```python
x = dataset.iloc[:,:-1]
y = dataset["MEDV"]
```

```python
print(x)
print(y)
```

#### Splitting Data into Training and Testing Data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2,random_state=42)
```

```python
print(x.shape,x_train.shape, x_test.shape)
```

#### Model Training

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
lr.score(x_test,y_test)*100
```

# Evaluation

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# Mean squared Error
print(mean_squared_error(y_train,lr.predict(x_train)))

#Mean Absolute Error
print(mean_absolute_error(y_train,lr.predict(x_train)))
```

```python
# Mean squared Error
print(mean_squared_error(y_test,lr.predict(x_test)))

#Mean Absolute Error
print(mean_absolute_error(y_test,lr.predict(x_test)))
```

#### Visualizing the actual Prices and Predicted prices

```python
#training Data Visualisation
plt.scatter(y_train, lr.predict(x_train))
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Predicted Price")
plt.show()
```

```python
#testing Data Visualisation
plt.scatter(y_test,lr.predict(x_test))
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Predicted Price")
plt.show()
```
