```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
dataset = pd.read_csv("car-price-dataset.csv")
```

### Handling missing values

```python
dataset["mileage(km/ltr/kg)"] = dataset["mileage(km/ltr/kg)"].fillna(dataset["mileage(km/ltr/kg)"].mean())
dataset["engine"] = dataset["engine"].fillna(dataset["engine"].mean())
dataset["max_power"] = dataset["max_power"].fillna(dataset["max_power"].mode()[0])
dataset["seats"] = dataset["seats"].fillna(dataset["seats"].mean())
```

```python
dataset.isnull().sum()
```

```python
dataset['max_power'].apply(lambda x: x.count(' ')).sum()
```

```python
dataset['max_power'] = dataset['max_power'].str.replace(' ', '0')
```

```python
dataset['max_power'] = dataset['max_power'].astype(float)
```

```python
dataset.select_dtypes(include=["object"])
```

### Label Encoding

```python
le = LabelEncoder()	
dataset["name"] = le.fit_transform(dataset["name"])
```

```python
dataset["fuel"] = le.fit_transform(dataset["fuel"])
dataset["seller_type"] = le.fit_transform(dataset["seller_type"])
dataset["transmission"] = le.fit_transform(dataset["transmission"])
dataset["owner"] = le.fit_transform(dataset["owner"])
```

```python
dataset.head()
```

```python
correlation = dataset.corr()
sns.heatmap(correlation, cbar=True,square=True,fmt='.2f',annot=True,annot_kws={'size':8},cmap = 'Blues')
plt.show()
```

### Modeling

```python
x = dataset.drop(columns = ['selling_price'])
y = dataset['selling_price']
print(f'size of input data {x.shape}')
print(f'size of output data {y.shape}')
```

#### train test spliting

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

#### Linear Regression

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
print(mean_squared_error(y_test,lr.predict(x_test)))
print(mean_absolute_error(y_test,lr.predict(x_test)))
print(np.sqrt(mean_squared_error(y_test,lr.predict(x_test))))
```

```python
plt.figure(figsize=(15,5))
plt.bar(x.columns,lr.coef_)
plt.title("Linear Regression")
plt.xlabel("Columns")
plt.ylabel("Coeff")
plt.show()
```

#### Lasso Regression

```python
ls = Lasso(alpha=0.5)
ls.fit(x_train, y_train)
```

```python
ls.score(x_test,y_test)*100
```

```python
print(mean_squared_error(y_test,ls.predict(x_test)))
print(mean_absolute_error(y_test,ls.predict(x_test)))
print(np.sqrt(mean_squared_error(y_test,ls.predict(x_test))))
```
