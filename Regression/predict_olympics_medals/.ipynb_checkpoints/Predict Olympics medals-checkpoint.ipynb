# Predict Olympics Medals

Use data from historical Olympic games.
predict how many medals a country will win based on historical and current data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("teams.csv")
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset = dataset[["team","country", "year","athletes","age","prev_medals","medals"]]
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset=dataset.dropna()
```

```python
dataset.isnull().sum()
```

```python
corr = dataset.select_dtypes(["int64", "float64"])
```

```python
corr.corr()["medals"]
```

```python
sns.lmplot(x="athletes",y="medals", data=dataset)
plt.show()
```

```python
train = dataset[dataset['year'] < 2012].copy()
test = dataset[dataset['year'] >= 2012].copy()
```

```python
train.shape
```

```python
test.shape
```

```python
input_data = ["athletes","prev_medals"]
output_data = ["medals"]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(train[input_data], train["medals"])
```

```python
lr.score(test[input_data], test["medals"])
```

```python
prd = lr.predict(test[input_data])
```

```python
test["predictions"] = prd
```

```python
test
```

```python
test.loc[test["predictions"] < 0, "predictions"] = 0
```

```python
test
```

```python
test["predictions"] = test["predictions"].round()
```

```python
test.head()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
print("Mean Squared Error: ", mean_squared_error(test["medals"], test["predictions"]))
print("Mean absolute Error: ", mean_absolute_error(test["medals"], test["predictions"]))
print("Mean Squared Error: ", np.sqrt(mean_squared_error(test["medals"], test["predictions"])))
```
