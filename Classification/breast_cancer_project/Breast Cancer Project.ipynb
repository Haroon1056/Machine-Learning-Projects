# Breast Cancer Project

## import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("data.csv")
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
sns.heatmap(dataset.isnull())
plt.show()
```

## Clean the data

```python
dataset.drop(columns=["id", "Unnamed: 32"], inplace=True)
```

```python
dataset.head()
```

```python
dataset["diagnosis"].replace("M", "1", inplace=True)
dataset["diagnosis"].replace("B", "0", inplace=True)
```

```python
dataset["diagnosis"] = dataset["diagnosis"].astype("int64")
```

```python
dataset.head()
```

```python
plt.figure(figsize=(4,3))
sns.barplot(dataset["diagnosis"].value_counts())
plt.show()
```

```python
x = dataset.drop(["diagnosis"], axis=1)
y = dataset["diagnosis"]
```

## normalize the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
x = ss.fit_transform(x)
```

```python
x
```

## split the data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25, random_state=42)
```

## train the model

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_prd = lr.predict(x_test)
```

```python
y_prd
```

## evaluation of the model 

```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score, accuracy_score, f1_score
```

```python
cf = confusion_matrix(y_test, y_prd)
```

```python
cf
```

```python
plt.figure(figsize=(4,3))
sns.heatmap(cf, annot=True)
plt.show()
```

```python
precision_score(y_test,y_prd)
```

```python
recall_score(y_test,y_prd)
```

```python
f1_score(y_test,y_prd)
```

```python
accuracy_score(y_test,y_prd)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_prd))
```
